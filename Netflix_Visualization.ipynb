```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv('https://raw.githubusercontent.com/Sofya27/Netflix-Visualization/main/Netflix%20TV%20Shows%20and%20Movies.csv')
```

```python
# 1 - seeing if release year influence the rating of the movie
netflix.plot(kind='scatter', x = 'release_year', y = 'imdb_score')
plt.show()
```

```python
# 2 - seeing if runtime (length of the movie) influnce the score
netflix.plot(kind='scatter', x = 'runtime', y = 'imdb_score')
plt.show()
```

```python
# 3 - checking release year and score from different perspective
sns.jointplot(x = 'release_year', y = 'imdb_score', data = netflix, size =5)
plt.show()
```

```python
# 4 - seing what ratings are for rantime and release year
sns.FacetGrid(netflix, hue='imdb_score', palette='husl').map(plt.scatter, 'runtime', 'release_year').add_legend()
plt.show()
```

```python
# 5 - seeing scores by type
sns.boxplot(x='type', y='imdb_score', palette='muted', data=netflix)
plt.show()
```

```python
# 6 - seeing scores by type
sns.boxplot(x='age_certification', y='imdb_score', palette='deep', data=netflix)
plt.show()
```

```python
# 7 - seing the density of scores for age restriction
sns.violinplot(x='age_certification', y='imdb_score', palette='Paired', data=netflix)
plt.show()
```

```python
# 8 - seing the density of scores for types
sns.violinplot(x='type', y='imdb_score', palette='mako', data=netflix)
#sns.swarmplot(x='type', y='imdb_score', data=netflix, color='w', alpha=.9);
plt.show()
```

```python
# 9 - pairplot of all numerical variables
sns.pairplot(netflix, hue='imdb_score', palette='mako')
plt.show()
```

```python
# 10 -  distribution of scores
netflix.boxplot(by='imdb_score', figsize=(12, 10))
plt.show()
```

```python
# 11 - distribution by scores, votes and runtime
import plotly.express as px #learned from https://medium.com/plotly/introducing-plotly-express-808df010143d
fig = px.parallel_coordinates(netflix[['imdb_score', 'imdb_votes', 'runtime']])
fig.show()
```

```python
# 12 - distribution of scores by release year
fig = px.box(netflix, x='release_year', y='imdb_score')
fig.show()
```

```python
# 13 - scores by age certification
fig = px.parallel_categories(netflix, dimensions=['age_certification', 'imdb_score'])
fig.show()
```

```python
# 14 - making a correlation matrix
sns.heatmap(netflix.corr())
plt.title('Correlation Plot')
plt.show()
```

```python
# 15 visualizing the distribution of type
data =  netflix.type.value_counts()
plt.figure(figsize=(5,10))
data.plot.pie( title = 'Pie Chart Showing Type Distribution')
plt.show()
```

```python
# 16 - age distribution by type
plt.figure(figsize=(10, 6))
sns.countplot(x='age_certification', hue='type', data=netflix)
plt.show()
```
